```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("C:/Users/zul_l/Downloads/wine.data")
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
predictors_col = ["Alcohol","AcidoMalico","Ceniza","AlcalinidadDeLACeniza","Magnesio","FenolesTotales","Flavonoides","FenolesnoFlavonoides","Proantocianinas","IntensidadDelColor","Tono","OD280","Prolina"]
target_col = ['Clase']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2,random_state=42)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
pd.crosstab(np.array([y[0]for y in y_test.values.tolist()]),predicciones,rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test,predicciones)
accuracy
```
